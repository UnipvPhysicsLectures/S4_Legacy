## py-matrix solution

```python
#------building the optical constant database------
eps_db_out=pm.mat.generate_eps_db('/home/pellegr/electrodynamics/st-matrix/py_matrix/materials/',ext='*.edb')
eps_files,eps_names,eps_db=eps_db_out['eps_files'],eps_db_out['eps_names'],eps_db_out['eps_db']
```

```python
stack=['e_bk7','e_au','e_vacuum'];
d_list=np.array([0.0,55.0,0.0]);
wl_0=633;
theta_min=40*np.pi/1.8e2;
theta_max=50*np.pi/1.8e2;
theta_step=num_cores*4;
v_theta=np.linspace(theta_min,theta_max,theta_step)
phi_0=0.0
theta_0=0.75172842585897504

# drude metal parameters
w_p=8.95 # plasma frequency
gamma=0.02 # damping constant
f_m=1.06e-6 # medium magnetic coupling
v_f=1.4e6; # Fermi velocity
gamma_inf=0.069; # bulk damping constant

# B field in Tesla
B=10.0
```

```python
# retrieving optical constants at wl from the database
e_list=np.array(pm.mat.db_to_eps(wl_0,eps_db,stack))
e_xy=pm.mat.eps_xy_drude(wl_0, w_p, gamma_inf, B)
e_list_off=np.zeros_like(e_list)
#e_list_off[1]=e_xy

# filling dielectric tensor diagonal
m_eps=np.zeros((len(stack),3,3),dtype=np.complex128);
m_eps[:,0,0]=e_list
m_eps[:,1,1]=e_list
m_eps[:,2,2]=e_list
m_eps_diag=m_eps.copy()

# filling off diagonal
m_eps[1,1,0]=e_xy
m_eps[1,0,1]=-e_xy
```

```python
# eps tensor and reflectance output vector
v_R_py=np.zeros_like(v_theta)
v_R_tm=np.zeros_like(v_theta)

# angle loop
for i_t,t in enumerate(v_theta):
    
    #------Computing------
    m_r_ps=pm.core.rt(wl_0,t,phi_0,m_eps,d_list)['m_r_ps']
    v_R_py[i_t]=pm.utils.R_ps_rl(m_r_ps)['R_p']
    
    #------Computing with mo subroutine------
    m_r_ps,error = tm.tmatrix.reflectance(wl_0,t,phi_0,e_list,e_list_off,d_list,"tmoke");
    v_R_tm[i_t]=pm.utils.R_ps_rl(m_r_ps)['R_p']
```

```python
# arguments
n_l=1;
# wavevector modulus and in plane components
n_0=np.sqrt(m_eps[0,0,0])
k0 = 2.0*np.pi/wl_0
kx = -k0*n_0*np.sin(theta_0)*np.cos(phi_0)
ky = -k0*n_0*np.sin(theta_0)*np.sin(phi_0)
```

```python
# kz eigenvalues for the anysotropic layer
v_kz = pm.core.kz_eigenvalues(k0,kx,ky,m_eps[n_l])
print 'v_kz'
print v_kz[0],v_kz[1],v_kz[2],v_kz[3]
```

```python
m=0
# k matrix
m_k=np.zeros_like(m_eps[n_l])
m_k[0,0] = 0.0;m_k[0,1] = -v_kz[m];m_k[0,2] = ky
m_k[1,0] = v_kz[m];m_k[1,1] = 0.0;m_k[1,2] = -kx
m_k[2,0] = -ky;m_k[2,1] = kx;m_k[2,2] = 0.0

# Characteristic matrix
m_char = np.dot(m_k,m_k)/(k0**2)
m_char = m_char+m_eps[n_l]
```
